```python
# -*- coding: utf-8 -*-
"""
Comprehensive Analysis of California Fire Incidents (2013-2019)

This script performs data loading, cleaning, feature engineering, model training
(regression and classification), evaluation, and visualization based on the
California Fire Incidents dataset from Kaggle. It aims to reproduce the
analysis described in the accompanying report.

Dataset Source: https://www.kaggle.com/datasets/ananthu017/california-wildfire-incidents-20132020
(Note: Analysis focused on 2013-2019 based on initial exploration)
"""

# @title 1. Setup: Import Libraries and Install Packages
# Install necessary libraries if running in Colab or similar environment
!pip install category_encoders lightgbm catboost missingno -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
import time
import warnings

# Modeling & Evaluation imports
from sklearn.model_selection import train_test_split, RandomizedSearchCV, KFold, cross_val_score
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.linear_model import Ridge, LogisticRegression
import lightgbm as lgb
from catboost import CatBoostRegressor, CatBoostClassifier
from category_encoders import TargetEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    mean_absolute_error, mean_squared_error, r2_score, # Regression
    accuracy_score, precision_score, recall_score, f1_score, # Classification
    roc_auc_score, confusion_matrix, ConfusionMatrixDisplay,
    classification_report
)
from scipy.stats import randint, uniform, loguniform # For RandomizedSearchCV


# Configure settings
pd.set_option('display.max_columns', None); pd.set_option('display.max_rows', 100)
pd.set_option('display.width', 1000)
warnings.filterwarnings('ignore', category=UserWarning)
warnings.filterwarnings('ignore', category=FutureWarning)
sns.set_style("whitegrid") # Set a plotting style

# --- Configuration Constants ---
FILE_PATH = '/content/California_Fire_Incidents 7 (1).csv' # Make sure this path is correct
RESULTS_DIR = '/content/results/' # Directory to save plots
RARE_CATEGORY_THRESHOLD = 10
TARGET_REGRESSION = 'AcresBurned_log'
TARGET_CLASSIFICATION = 'RiskLevel_1000'
RISK_THRESHOLD_ACRES = 1000
N_SPLITS_CV = 5
N_ITER_TUNING = 50 # Iterations for RandomizedSearch
RANDOM_STATE = 42 # For reproducibility

# Create results directory if it doesn't exist
import os
os.makedirs(RESULTS_DIR, exist_ok=True)

print("Setup Complete.")

# @title 2. Load Data
# --- Load Data ---
try:
    df_raw = pd.read_csv(FILE_PATH)
    print(f"Successfully loaded '{FILE_PATH}' ({len(df_raw)} rows)\n")
except FileNotFoundError:
    print(f"Error: File not found at '{FILE_PATH}'. Please upload the file to Colab session storage.")
    # Optional: Add code here to upload file if in Colab
    # from google.colab import files
    # print("Please upload the CSV file:")
    # uploaded = files.upload()
    # if FILE_PATH.split('/')[-1] in uploaded:
    #     df_raw = pd.read_csv(FILE_PATH.split('/')[-1])
    #     print(f"Successfully loaded '{FILE_PATH.split('/')[-1]}'")
    # else:
    #     print("Upload failed or file name mismatch.")
    #     exit()
    exit() # Exit if file still not found
except Exception as e:
    print(f"An error occurred loading the CSV: {e}"); exit()

# Basic Inspection
print("--- Initial Data Head ---")
display(df_raw.head())
print("\n--- Initial Data Info ---")
df_raw.info()

# @title 3. Exploratory Data Analysis (EDA) & Initial Cleaning

# --- Standardize Column Names ---
df_raw.columns = df_raw.columns.str.strip().str.lower().str.replace(' ', '_')
print("\n--- Standardized Column Names ---")
print(df_raw.columns)

# --- Missing Value Analysis (Figure 1) ---
print("\n--- Generating Missing Value Plot (Figure 1) ---")
missing_percent = (df_raw.isnull().sum() / len(df_raw)) * 100
missing_percent_filtered = missing_percent[missing_percent > 0].sort_values(ascending=True)

plt.figure(figsize=(10, 12))
bars = plt.barh(missing_percent_filtered.index, missing_percent_filtered.values, color='skyblue')
plt.xlabel('Percentage Missing (%)', fontsize=12)
plt.ylabel('Dataset Columns', fontsize=12)
plt.title('Figure 1: Percentage of Missing Values per Column\n(Highlighting Data Scarcity for Resource/Impact Variables)', fontsize=14, fontweight='bold')
plt.xticks(fontsize=10); plt.yticks(fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.6)
for bar in bars:
    width = bar.get_width()
    plt.text(width + 1, bar.get_y() + bar.get_height()/2, f'{width:.1f}%', va='center', fontsize=9)
plt.tight_layout()
figure_save_path = os.path.join(RESULTS_DIR, 'figure1_missing_values.png')
plt.savefig(figure_save_path, dpi=300)
print(f"Figure 1 saved to '{figure_save_path}'")
plt.show()

# Display Missing Value Table
print("\n--- Missing Value Percentages (Full) ---")
print(missing_percent.sort_values(ascending=False))

# --- Target Variable Distribution ---
print("\n--- Target Variable (AcresBurned) Distribution ---")
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df_raw['acresburned'].dropna(), bins=50, kde=False)
plt.title('Original AcresBurned Distribution')
plt.xlabel('Acres Burned')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
# Use log1p for log transform (handles 0 acres)
sns.histplot(np.log1p(df_raw['acresburned'].dropna()), bins=50, kde=True)
plt.title('Log1p(AcresBurned) Distribution')
plt.xlabel('log(Acres Burned + 1)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
print("Note the extreme right skew in the original distribution, justifying log transformation for regression.")

# @title 4. Preprocessing and Feature Engineering Function
# --- Base Cleaning & Feature Engineering Function ---
def clean_and_engineer(df_in, target_variable_name, drop_original_acres=True):
    """
    Cleans data, engineers features, handles coordinates (marks as NaN),
    and prepares data structure, keeping original AcresBurned for target creation.
    """
    df_copy = df_in.copy()
    print("--- Starting Cleaning & Feature Engineering ---")

    # 1. Drop High Missing / Irrelevant Columns
    # Calculate missing based on input df
    missing_counts = df_copy.isnull().sum()
    missing_percentages = (missing_counts / len(df_copy)) * 100
    missing_df = pd.DataFrame({'Count': missing_counts, 'Percentage': missing_percentages})
    missing_threshold = 80
    high_missing_cols = missing_df[missing_df['Percentage'] > missing_threshold].index.tolist()

    other_cols_to_drop = [
         'uniqueid', 'canonicalurl', 'name', 'conditionstatement',
        'controlstatement', 'countyids', 'percentcontained', 'searchdescription',
        'searchkeywords', 'updated', 'extinguished', 'airtankers', 'crewsinvolved',
        'dozers', 'engines', 'fatalities', 'helicopters', 'injuries',
        'personnelinvolved', 'structuresdamaged', 'structuresdestroyed',
        'structuresevacuated', 'structuresthreatened', 'watertenders', 'fueltype',
        # Low importance/variance bools from prior analysis
        'active', 'featured', 'final', 'public'
    ]

    # Ensure crucial columns are not accidentally dropped yet
    keep_cols = ['started', 'latitude', 'longitude', 'acresburned', target_variable_name] # Keep original acres temporarily
    for col in keep_cols:
        if col in high_missing_cols: high_missing_cols.remove(col)
        if col in other_cols_to_drop: other_cols_to_drop.remove(col)

    cols_to_drop = list(set(high_missing_cols + other_cols_to_drop))
    if 'started' in cols_to_drop: cols_to_drop.remove('started') # Keep 'started' for FE

    # Perform the drop
    cols_present_to_drop = [col for col in cols_to_drop if col in df_copy.columns]
    df_processed = df_copy.drop(columns=cols_present_to_drop)
    print(f"Dropped {len(cols_present_to_drop)} columns. Kept: {list(df_processed.columns)}")

    # 2. Date/Time Features
    if 'started' not in df_processed.columns: raise ValueError("'started' column missing")
    df_processed['started'] = pd.to_datetime(df_processed['started'], errors='coerce')
    original_rows = len(df_processed)
    df_processed.dropna(subset=['started'], inplace=True)
    print(f"Dropped {original_rows - len(df_processed)} rows with invalid 'started' date.")

    df_processed['Start_Year'] = df_processed['started'].dt.year
    # df_processed['Start_Month'] = df_processed['started'].dt.month # Using cyclical
    df_processed['Start_DayOfWeek'] = df_processed['started'].dt.dayofweek
    df_processed['Start_DayOfYear'] = df_processed['started'].dt.dayofyear
    # Cyclical Features
    df_processed['Month_sin'] = np.sin(2 * np.pi * df_processed['started'].dt.month/12)
    df_processed['Month_cos'] = np.cos(2 * np.pi * df_processed['started'].dt.month/12)
    df_processed['DayOfYear_sin'] = np.sin(2 * np.pi * df_processed['Start_DayOfYear']/365.25)
    df_processed['DayOfYear_cos'] = np.cos(2 * np.pi * df_processed['Start_DayOfYear']/365.25)

    # Drop original/intermediate date columns
    cols_to_drop_dates = ['started', 'archiveyear', 'Start_DayOfYear'] # Dropping DayOfYear too now
    df_processed = df_processed.drop(columns=[col for col in cols_to_drop_dates if col in df_processed.columns])
    print("Engineered date features (incl. cyclical).")

    # 3. Clean Coordinates (Mark Invalid as NaN)
    if 'latitude' not in df_processed.columns or 'longitude' not in df_processed.columns:
        raise ValueError("Coordinate columns missing")
    df_processed['latitude'] = df_processed['latitude'].replace(0.0, np.nan)
    df_processed['longitude'] = df_processed['longitude'].replace(0.0, np.nan)
    lat_min, lat_max = 32.0, 42.5; lon_min, lon_max = -125.0, -114.0
    df_processed['latitude'] = df_processed['latitude'].apply(lambda x: x if pd.isna(x) or (lat_min <= x <= lat_max) else np.nan)
    df_processed['longitude'] = df_processed['longitude'].apply(lambda x: x if pd.isna(x) or (lon_min <= x <= lon_max) else np.nan)
    print(f"Marked invalid coordinates as NaN. Missing Lat/Lon: {df_processed[['latitude', 'longitude']].isnull().sum().sum()}")

    # 4. Drop Location (High Cardinality) & Status (Low Importance)
    if 'location' in df_processed.columns: df_processed = df_processed.drop(columns=['location'])
    if 'status' in df_processed.columns: df_processed = df_processed.drop(columns=['status'])

    # 5. Convert remaining bools to int
    boolean_features = df_processed.select_dtypes(include='bool').columns.tolist()
    for col in boolean_features:
        df_processed[col] = df_processed[col].astype(int)
    if boolean_features: print(f"Converted booleans to int: {boolean_features}")

    # 6. Handle original AcresBurned based on flag
    if drop_original_acres and 'acresburned' in df_processed.columns:
        df_processed = df_processed.drop(columns=['acresburned'])
        print("Dropped original 'acresburned' column.")

    return df_processed

# @title 5. Regression Analysis: Predicting log(AcresBurned)

# --- Prepare Data for Regression ---
print("\n" + "="*20 + " Regression Analysis " + "="*20)
df_reg = df_raw.copy() # Start fresh for this analysis
df_reg.dropna(subset=['acresburned'], inplace=True) # Drop rows where original acres is missing
df_reg[TARGET_REGRESSION] = np.log1p(df_reg['acresburned'])
df_cleaned_reg = clean_and_engineer(df_reg, TARGET_REGRESSION, drop_original_acres=True)

# --- Strategy Comparison for Coordinate Handling ---
# Strategy 1R: Drop Missing Coords
df_drop_reg = df_cleaned_reg.copy()
original_rows_reg = len(df_drop_reg)
df_drop_reg.dropna(subset=['latitude', 'longitude'], inplace=True)
print(f"\nStrategy 1R (Drop): Dropped {original_rows_reg - len(df_drop_reg)} rows with missing coords.")
y_drop_reg = df_drop_reg[TARGET_REGRESSION]
X_drop_reg = df_drop_reg.drop(columns=[TARGET_REGRESSION])

# Strategy 2R: KNN Impute Missing Coords
df_knn_reg = df_cleaned_reg.copy()
y_knn_reg = df_knn_reg[TARGET_REGRESSION]
X_knn_reg = df_knn_reg.drop(columns=[TARGET_REGRESSION])
print(f"\nStrategy 2R (KNN Impute): Dataset size {len(X_knn_reg)}")

# --- Define Feature Types (using drop schema as reference) ---
numerical_features_reg = X_drop_reg.select_dtypes(include=np.number).columns.tolist()
categorical_features_reg = X_drop_reg.select_dtypes(include='object').columns.tolist()

# --- Regression Preprocessing Pipelines ---
numeric_pipeline_knn_reg = Pipeline(steps=[
    ('scaler', StandardScaler()),
    ('imputer_knn', KNNImputer(n_neighbors=5))
])
numeric_pipeline_drop_reg = SimpleImputer(strategy='median')
categorical_pipeline_reg = Pipeline(steps=[
    ('imputer_cat', SimpleImputer(strategy='constant', fill_value='__MISSING__')),
    ('target_encoder', TargetEncoder(smoothing=10.0, min_samples_leaf=5, handle_unknown='value', handle_missing='value'))
])

# --- Function to Evaluate Regression ---
def evaluate_regression_strategy(X, y, numeric_transformer, strategy_name):
    print(f"\n===== Evaluating Regression Strategy: {strategy_name} =====")
    # Define types based on input X
    current_numerical = X.select_dtypes(include=np.number).columns.tolist()
    current_categorical = X.select_dtypes(include='object').columns.tolist()

    # Group rare categories
    X_processed = X.copy()
    for col in current_categorical:
        value_counts = X_processed[col].value_counts()
        rare_categories = value_counts[value_counts < RARE_CATEGORY_THRESHOLD].index
        if len(rare_categories) > 0:
            X_processed[col] = X_processed[col].replace(rare_categories, f'Other_{col}')

    # Preprocessor
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, current_numerical),
            ('cat', categorical_pipeline_reg, current_categorical)
        ], remainder='passthrough'
    )

    X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=RANDOM_STATE)
    print(f"Train/Test Split: {len(X_train)} / {len(X_test)}")

    results_reg = {}
    models_reg = {
        "Ridge": Ridge(random_state=RANDOM_STATE),
        "RandomForest": RandomForestRegressor(random_state=RANDOM_STATE, n_jobs=-1, n_estimators=100),
        "CatBoost": CatBoostRegressor(random_state=RANDOM_STATE, verbose=0, allow_writing_files=False)
    }
    param_grids_reg = {
        "Ridge": {'regressor__alpha': loguniform(1e-2, 1e3)},
        "RandomForest": {'regressor__n_estimators': randint(100, 500), 'regressor__max_depth': randint(5, 25), 'regressor__min_samples_leaf': randint(1, 15)},
        "CatBoost": {'regressor__iterations': randint(100, 1000), 'regressor__learning_rate': loguniform(0.01, 0.2), 'regressor__depth': randint(4, 10)}
    }

    cv_strategy = KFold(n_splits=N_SPLITS_CV, shuffle=True, random_state=RANDOM_STATE)

    for name, model in models_reg.items():
        print(f"--- Tuning/Evaluating {name} for {strategy_name} ---")
        start_time = time.time()
        pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('regressor', model)])
        random_search = RandomizedSearchCV(
            pipeline, param_distributions=param_grids_reg[name], n_iter=N_ITER_TUNING if name != "Ridge" else 15,
            cv=cv_strategy, scoring='r2', n_jobs=-1, random_state=RANDOM_STATE, verbose=0
        )
        random_search.fit(X_train, y_train)
        best_model = random_search.best_estimator_
        y_pred_log = best_model.predict(X_test)

        r2 = r2_score(y_test, y_pred_log)
        mae_log = mean_absolute_error(y_test, y_pred_log)
        rmse_log = np.sqrt(mean_squared_error(y_test, y_pred_log))
        y_test_orig = np.expm1(y_test); y_pred_orig = np.expm1(y_pred_log)
        y_pred_orig[y_pred_orig < 0] = 0
        mae_orig = mean_absolute_error(y_test_orig, y_pred_orig)

        results_reg[name] = {
            'Best CV R2': random_search.best_score_, 'Test R2': r2, 'Test MAE (log)': mae_log,
            'Test RMSE (log)': rmse_log, 'Test MAE (acres)': mae_orig, 'Best Params': random_search.best_params_,
            'Fit Time (s)': time.time() - start_time, 'Model Object': best_model
        }
        print(f"  {name} - Test R2: {r2:.4f}, MAE (acres): {mae_orig:.2f}, Time: {results_reg[name]['Fit Time (s)']:.1f}s")

    return results_reg

# --- Run Regression Evaluations ---
results_drop_reg = evaluate_regression_strategy(X_drop_reg, y_drop_reg, numeric_pipeline_drop_reg, "Drop Missing Coords")
results_knn_reg = evaluate_regression_strategy(X_knn_reg, y_knn_reg, numeric_pipeline_knn_reg, "KNN Impute Coords")

# --- Regression Summary Table ---
print("\n--- Regression Results Summary ---")
summary_list_reg = []
for name, metrics in results_drop_reg.items():
    summary_list_reg.append({'Strategy': 'Drop', 'Model': name, 'Test R2': metrics['Test R2'], 'MAE (acres)': metrics['Test MAE (acres)']})
for name, metrics in results_knn_reg.items():
    summary_list_reg.append({'Strategy': 'KNN Impute', 'Model': name, 'Test R2': metrics['Test R2'], 'MAE (acres)': metrics['Test MAE (acres)']})
summary_df_reg = pd.DataFrame(summary_list_reg).sort_values(by='Test R2', ascending=False)
print(summary_df_reg)
best_reg_result = summary_df_reg.iloc[0]

# @title 6. Classification Analysis: Predicting RiskLevel > 1000 Acres

# --- Prepare Data for Classification ---
print("\n" + "="*20 + " Classification Analysis " + "="*20)
df_cls = df_raw.copy() # Start fresh
df_cleaned_cls = clean_and_engineer(df_cls, target_variable_name='acresburned', drop_original_acres=False) # Keep acresburned to create target

# Create Binary Target
df_cleaned_cls[TARGET_CLASSIFICATION] = (df_cleaned_cls['acresburned'] > RISK_THRESHOLD_ACRES).astype(int)
print(f"\nCreated target '{TARGET_CLASSIFICATION}' (1 if AcresBurned > {RISK_THRESHOLD_ACRES}, else 0)")

# Drop rows with missing coordinates for this analysis
original_rows_cls = len(df_cleaned_cls)
df_cleaned_cls.dropna(subset=['latitude', 'longitude'], inplace=True)
print(f"Dropped {original_rows_cls - len(df_cleaned_cls)} rows with missing coords for classification.")

# Check Class Imbalance
print("\n--- Target Class Distribution ---")
class_counts = df_cleaned_cls[TARGET_CLASSIFICATION].value_counts()
class_percentages = df_cleaned_cls[TARGET_CLASSIFICATION].value_counts(normalize=True) * 100
print(class_counts)
print(f"\nLow Risk (0): {class_percentages.get(0, 0):.2f}%")
print(f"High Risk (1): {class_percentages.get(1, 0):.2f}%")
if class_percentages.min() < 15:
    print("WARNING: Significant class imbalance detected.")
    SCORING_METRIC = 'roc_auc'
else:
    SCORING_METRIC = 'f1_weighted' # F1-weighted might be better than accuracy if slightly imbalanced

# Prepare final X, y for classification
y_cls = df_cleaned_cls[TARGET_CLASSIFICATION]
X_cls = df_cleaned_cls.drop(columns=[TARGET_CLASSIFICATION, 'acresburned']) # Drop target AND original acres
print("\nSeparated final features (X_cls) and target (y_cls).")

# --- Group Rare Categories & Identify Feature Types ---
print("\nGrouping Rare Categories for Classification...")
categorical_features_cls = X_cls.select_dtypes(include='object').columns.tolist()
for col in categorical_features_cls:
    value_counts = X_cls[col].value_counts()
    rare_categories = value_counts[value_counts < RARE_CATEGORY_THRESHOLD].index
    if len(rare_categories) > 0:
        print(f"Grouping {len(rare_categories)} rare categories in '{col}'")
        X_cls[col] = X_cls[col].replace(rare_categories, f'Other_{col}')

numerical_features_cls = X_cls.select_dtypes(include=np.number).columns.tolist()
categorical_features_cls = X_cls.select_dtypes(include='object').columns.tolist() # Re-identify after grouping
print(f"Final Numerical Features: {numerical_features_cls}")
print(f"Final Categorical Features: {categorical_features_cls}")
print(f"Final feature shape: {X_cls.shape}")

# --- Train/Test Split (Stratified) ---
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X_cls, y_cls, test_size=0.2, random_state=RANDOM_STATE, stratify=y_cls
)
print(f"\nSplit classification data into {len(X_train_cls)} training and {len(X_test_cls)} testing samples (stratified).")

# --- Classification Preprocessing Pipeline ---
# Use median imputer for numerical, target encoder for categorical
numeric_pipeline_cls = SimpleImputer(strategy='median')
categorical_pipeline_cls = Pipeline(steps=[
    ('imputer_cat', SimpleImputer(strategy='constant', fill_value='__MISSING__')),
    ('target_encoder', TargetEncoder(smoothing=10.0, min_samples_leaf=5, handle_unknown='value', handle_missing='value'))
])

preprocessor_cls = ColumnTransformer(
    transformers=[
        ('num', numeric_pipeline_cls, numerical_features_cls),
        ('cat', categorical_pipeline_cls, categorical_features_cls)
    ],
    remainder='passthrough'
)

# --- Function to Evaluate Classification ---
def evaluate_classification_strategy(X_train, X_test, y_train, y_test, preprocessor, model_name, model_instance, param_grid):
    print(f"\n===== Evaluating Classification Model: {model_name} =====")
    start_time = time.time()
    pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', model_instance)])
    cv_strategy = KFold(n_splits=N_SPLITS_CV, shuffle=True, random_state=RANDOM_STATE)

    random_search = RandomizedSearchCV(
        pipeline, param_distributions=param_grid, n_iter=N_ITER_TUNING,
        cv=cv_strategy, scoring=SCORING_METRIC, n_jobs=-1,
        random_state=RANDOM_STATE, verbose=0
    )
    random_search.fit(X_train, y_train)
    best_model = random_search.best_estimator_
    y_pred = best_model.predict(X_test)
    y_pred_proba = best_model.predict_proba(X_test)[:, 1] # Prob for class 1

    # Metrics
    accuracy = accuracy_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    f1_w = f1_score(y_test, y_pred, average='weighted')
    try:
        # Use zero_division=0 to handle cases where precision/recall might be 0
        class_report = classification_report(y_test, y_pred, target_names=['Low Risk (0)', 'High Risk (1)'], output_dict=True, zero_division=0)
        precision_hr = class_report['High Risk (1)']['precision']
        recall_hr = class_report['High Risk (1)']['recall']
        f1_hr = class_report['High Risk (1)']['f1-score']
    except Exception as e:
        print(f"Warning: Could not calculate detailed classification report: {e}")
        precision_hr, recall_hr, f1_hr = np.nan, np.nan, np.nan


    cm = confusion_matrix(y_test, y_pred)

    results = {
        'Best CV Score': random_search.best_score_, 'Test Accuracy': accuracy,
        'Test ROC AUC': roc_auc, 'Test F1 Weighted': f1_w,
        'Precision (High Risk)': precision_hr, 'Recall (High Risk)': recall_hr,
        'F1 (High Risk)': f1_hr, 'Confusion Matrix': cm,
        'Best Params': random_search.best_params_, 'Fit Time (s)': time.time() - start_time,
        'Model Object': best_model
    }
    print(f"  Best CV {SCORING_METRIC.upper()}: {results['Best CV Score']:.4f}")
    print(f"  Test Accuracy: {accuracy:.4f}")
    print(f"  Test ROC AUC: {roc_auc:.4f}")
    print(f"  Recall (High Risk): {results['Recall (High Risk)']:.4f}")
    print(f"  Precision (High Risk): {results['Precision (High Risk)']:.4f}")
    print(f"  F1 (High Risk): {results['F1 (High Risk)']:.4f}")
    print(f"  Fit Time: {results['Fit Time (s)']:.1f}s")
    return results

# --- Define Classification Models and Grids ---
models_cls = {
    "LogisticRegression": LogisticRegression(random_state=RANDOM_STATE, max_iter=1000, class_weight='balanced'),
    "RandomForest": RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1, class_weight='balanced'),
    "LightGBM": lgb.LGBMClassifier(random_state=RANDOM_STATE, n_jobs=-1, class_weight='balanced'),
    "CatBoost": CatBoostClassifier(random_state=RANDOM_STATE, verbose=0, allow_writing_files=False, auto_class_weights='Balanced')
}
param_grids_cls = {
    "LogisticRegression": {'classifier__C': loguniform(1e-3, 1e3), 'classifier__solver': ['liblinear', 'saga']},
    "RandomForest": {'classifier__n_estimators': randint(100, 600), 'classifier__max_depth': randint(5, 25), 'classifier__min_samples_leaf': randint(1, 15)},
    "LightGBM": {'classifier__n_estimators': randint(100, 1000), 'classifier__learning_rate': loguniform(0.01, 0.2), 'classifier__num_leaves': randint(15, 50)},
    "CatBoost": {'classifier__iterations': randint(100, 1000), 'classifier__learning_rate': loguniform(0.01, 0.2), 'classifier__depth': randint(4, 10)}
}

# --- Run Classification Evaluations ---
all_results_cls = {}
for name, model in models_cls.items():
    all_results_cls[name] = evaluate_classification_strategy(
        X_train_cls, X_test_cls, y_train_cls, y_test_cls,
        preprocessor_cls, name, model, param_grids_cls[name]
    )

# --- Classification Summary Table ---
print("\n--- Classification Results Summary ---")
summary_list_cls = []
for name, metrics in all_results_cls.items():
    summary_list_cls.append({
        'Model': name, 'Test ROC AUC': metrics['Test ROC AUC'], 'Test Accuracy': metrics['Test Accuracy'],
        'F1 (High Risk)': metrics['F1 (High Risk)'], 'Recall (High Risk)': metrics['Recall (High Risk)'],
        'Precision (High Risk)': metrics['Precision (High Risk)']
    })
summary_df_cls = pd.DataFrame(summary_list_cls).sort_values(by='Test ROC AUC', ascending=False)
print(summary_df_cls)
best_cls_result = summary_df_cls.iloc[0]

# @title 7. Visualization of Best Classification Model Results

# --- Analysis of Best Classification Model ---
if not summary_df_cls.empty:
    best_model_name_cls = best_cls_result['Model']
    best_model_metrics_cls = all_results_cls[best_model_name_cls]
    best_pipeline_object_cls = best_model_metrics_cls['Model Object']

    print(f"\n--- Detailed Analysis for Best Classification Model: {best_model_name_cls} ---")
    print(f"  Best Parameters Found: {best_model_metrics_cls['Best Params']}")

    # Re-predict to ensure report consistency
    y_pred_best_cls = best_pipeline_object_cls.predict(X_test_cls)
    print("\n  Test Set Classification Report:")
    print(classification_report(y_test_cls, y_pred_best_cls, target_names=['Low Risk (0)', 'High Risk (1)'], zero_division=0))

    # --- Confusion Matrix (Figure 2) ---
    print("\n  Generating Confusion Matrix Plot (Figure 2) ---")
    cm = best_model_metrics_cls['Confusion Matrix']
    fig_cm, ax_cm = plt.subplots(figsize=(6, 6))
    cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Low Risk (0)', 'High Risk (1)'])
    cm_display.plot(cmap=plt.cm.Blues, ax=ax_cm)
    ax_cm.set_title(f'Figure 2: Confusion Matrix for {best_model_name_cls}')
    plt.tight_layout()
    cm_save_path = os.path.join(RESULTS_DIR, 'figure2_confusion_matrix.png')
    plt.savefig(cm_save_path, dpi=300)
    print(f"Figure 2 saved to '{cm_save_path}'")
    plt.show()

    # --- Feature Importance (Figure 3) ---
    print(f"\n--- Feature Importances for Best Model ({best_model_name_cls}) ---")
    try:
        fitted_preprocessor = best_pipeline_object_cls.named_steps['preprocessor']
        fitted_classifier = best_pipeline_object_cls.named_steps['classifier']

        # Get feature names after transformation by the fitted preprocessor
        # This requires careful handling depending on the encoders used in the *best* pipeline
        # Using get_feature_names_out if possible (requires newer sklearn)
        try:
            pipeline_feature_names = fitted_preprocessor.get_feature_names_out()
            # Clean up names generated by pipeline (remove prefixes like 'num__', 'cat__')
            cleaned_feature_names = [name.split('__')[-1] for name in pipeline_feature_names]
        except AttributeError:
            # Fallback for older sklearn or if get_feature_names_out fails
             print("Warning: Could not use get_feature_names_out. Using original feature lists - names might not perfectly match transformed columns.")
             num_features_in = X_train_cls.select_dtypes(include=np.number).columns.tolist()
             cat_features_in = X_train_cls.select_dtypes(include='object').columns.tolist()
             cleaned_feature_names = num_features_in + cat_features_in


        if hasattr(fitted_classifier, 'feature_importances_'):
            importances = fitted_classifier.feature_importances_
        elif hasattr(fitted_classifier, 'coef_'): # For Logistic Regression
            importances = np.abs(fitted_classifier.coef_[0])
        else:
            importances = None

        if importances is not None and len(cleaned_feature_names) == len(importances):
            importance_df = pd.DataFrame({
                'Feature': cleaned_feature_names,
                'Importance': importances
            }).sort_values(by='Importance', ascending=False)
            print(importance_df.head(20))

            plt.figure(figsize=(10, 8))
            sns.barplot(x="Importance", y="Feature", data=importance_df.head(15))
            plt.title(f"Figure 3: Feature Importance (Model: {best_model_name_cls})")
            plt.tight_layout()
            fi_save_path = os.path.join(RESULTS_DIR, 'figure3_feature_importance.png')
            plt.savefig(fi_save_path, dpi=300)
            print(f"Figure 3 saved to '{fi_save_path}'")
            plt.show()
        else:
            print(f"Warning: Could not generate feature importances for {best_model_name_cls} (length mismatch or unsupported attribute).")
            print(f"Feature names length: {len(cleaned_feature_names)}, Importances length: {len(importances) if importances is not None else 'N/A'}")


    except Exception as e:
        print(f"Could not extract or display feature importances: {e}")
else:
    print("Error: No classification models were successfully evaluated to show best results.")


print("\n--- Analysis Complete ---")
print(f"Plots saved in '{RESULTS_DIR}'")
```
